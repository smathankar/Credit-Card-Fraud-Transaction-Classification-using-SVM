```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import numpy as np
import pandas as pd
%matplotlib inline



#dataset = pd.read_csv('creditcard.csv', sep=',',header=None)
dataset = pd.read_csv('creditcard.csv')
```

```python

#dataset.count()
#dataset.describe()
#dataset.info()

ds = dataset
```

```python
dataset.head(5)
```

```python
ds.head(5)
```

```python
class_label = pd.DataFrame(ds.iloc[:,-1])
class_label.columns = ['label']
ds = ds.iloc[:, :-1]
```

```python
ds.head(n=5)
```

```python
class_label.head(n=5)
```

```python


#Normalizing the data

ds = ds.sub(ds.mean(axis=0), axis=1)
```

```python
#Calculating covariance

df_mat = np.asmatrix(ds)
sigma = np.cov(df_mat.T)

#We have taken df_mat.T, as we are finding covariance among features of the data set.
#sigma will be our resultant covariance matrix with 4X4 dimensions.
```

```python
#Finding eigen values and eigen vectors


eigVals, eigVec = np.linalg.eig(sigma)
```

```python
#Sort the eigen values and eigen vectors

sorted_index = eigVals.argsort()[::-1]
eigVals = eigVals[sorted_index]
eigVec = eigVec[:,sorted_index]
```

```python
#Select top k eigen values and corresponding eigen vectors

eigVec = eigVec[:,:2]
```

```python
#Forming the new data set in reduced dimensions

transformed = df_mat.dot(eigVec)
transformed
```

```python
#horizontally stack transformed data set with class label.
final_df = np.hstack((transformed, class_label))
#convert the numpy array to data frame
final_df = pd.DataFrame(final_df)
#define the column names
final_df.columns = ['x','y','label']
```

```python
final_df.head(n=5)
```

```python
dataset.tail(5)
```

```python
#Plotting the new data set

groups = final_df.groupby('label')
figure, axes = plt.subplots()
axes.margins(0.05)
for name, group in groups:
    axes.plot(group.x, group.y, marker='o', linestyle='', ms=6, label=name)
    axes.set_title("PCA     0 = 'Non Fraudulant', 1 = 'Fraudulant'")
axes.legend()
plt.xlabel("principal component 1")
plt.ylabel("principal component 2")
plt.show()
```

# SVM with SckitLearn

```python
fdf = pd.DataFrame(final_df)
```

```python
fdf = fdf[(fdf.label == 1.0)].append(fdf.tail(50000))
#fdf = fdf[(fdf.label == 1.0)].append(fdf.tail(10000))
fdf.info()
```

```python
from sklearn.model_selection import train_test_split as tts
#array = np.array(fdf)
X = fdf[['x', 'y']]
Y = fdf[['label']]
seed = 10

X_train, X_test, Y_train, Y_test = tts(X, Y, test_size=0.20, random_state=seed)
```

```python
from sklearn.svm import SVC

svm = SVC(C=1.0, kernel='rbf', degree=2, gamma="auto", random_state=None)
#svm = SVC(C=1.0, kernel='linear')
model = svm.fit(X_train, Y_train)
model
Y_prediction = svm.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(confusion_matrix(Y_test, Y_prediction))
print(classification_report(Y_test, Y_prediction))
plt.matshow(confusion_matrix(Y_test, Y_prediction))
plt.colorbar()
```
